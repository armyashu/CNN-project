##### Change Dimention

```python
X_train=np.expand_dims(X_train,axis=-1)
```

```python
X_train.ndim
```

```python
X_train.shape
```

```python
X_test=np.expand_dims(X_test,axis=-1)
```

```python
X_test.ndim
```

```python
X_test.shape
```

##### Feature scaling

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

```python
X_test[0]
```

#### spliting dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_validation,y_train,y_validation=train_test_split(X_train,Y_train,test_size=0.2,random_state=2020)
```

```python
X_train.shape,y_train.shape
```

```python
X_validation.shape,y_validation.shape
```

### # Built First Convalutional Neural Network

```python
# Convalutional neural network -Model Building
```

```python
# Building CNN model
```

```python
cnn_model = keras.models.Sequential([
                         keras.layers.Conv2D(filters=32, kernel_size=3, strides=(1,1), padding='valid',activation= 'relu', input_shape=[28,28,1]),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Flatten(),
                         keras.layers.Dense(units=128, activation='relu'),
                         keras.layers.Dense(units=10, activation='softmax')
])
                       
                                            
                                            
                                            
                                            
                                        
```

```python
cnn_model.summary()
```

```python
# complie the model
```

```python
cnn_model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

#### Train CNN model

```python
cnn_model.fit(X_train,y_train,epochs=10,batch_size=512, verbose=1, validation_data=(X_validation,y_validation))
```

#### test and evaluate model}

```python
#1 test and evalute model
# test the model
cnn_model.predict(np.expand_dims(X_test[0],axis=0)).round(2)
```

```python
np.argmax(cnn_model.predict(np.expand_dims(X_test[0],axis=0)).round(2))
```

```python
y_pred=cnn_model.predict(X_test).round(2)
y_pred
```

```python
Y_test[0]
```

```python
cnn_model.evaluate(X_test,Y_test)
```

#### Visualize output

```python
plt.figure(figsize=(16,16))
j=1
for i in np.random.randint(0,1000,60):
    plt.subplot(10,6,j);j+=1
    plt.imshow(X_test[i].reshape(28,28),cmap='Greys')
    plt.axis('off')
    plt.title('Actual={}  /  {} \nPredicted={} / {}'.format(class_labels[Y_test[i]],Y_test[i],class_labels[np.argmax(y_pred[i])],np.argmax(y_pred[i])))
    plt.subplots_adjust(wspace=0.5, hspace=0.5)
           
```

####  Confusion Matrix

```python
"""## Confusion Matrix"""
 
from sklearn.metrics import confusion_matrix
 
plt.figure(figsize=(16,9))
y_pred_labels = [ np.argmax(label) for label in y_pred ]
cm = confusion_matrix(Y_test, y_pred_labels)
 
# show cm 
sns.heatmap(cm, annot=True, fmt='d',xticklabels=class_labels, yticklabels=class_labels)
```

```python
# classification report 
from sklearn.metrics import classification_report
cr= classification_report(Y_test, y_pred_labels, target_names=class_labels)
```
